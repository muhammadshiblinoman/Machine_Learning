```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.linear_model as linear_model
```

```python
df = pd.read_csv("capita_income.csv")
df.head(5)
```

```python
%matplotlib inline
plt.scatter(df.year, df.per_capita_income, color='blue', marker='o')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Per Capita Income Over Years')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```
