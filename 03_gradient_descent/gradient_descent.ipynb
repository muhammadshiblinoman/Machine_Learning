```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(x, y):
    m_curr = b_curr = 0
    rate = 0.01
    n = len(x)

    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='red', marker='+', linewidth=2)

    for i in range(100000):
        y_predicted = m_curr * x + b_curr
        if i % 10 == 0:
            plt.plot(x, y_predicted, color='green', alpha=0.3)
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * bd
    plt.plot(x, m_curr * x + b_curr, color='blue', linewidth=2, label='Final Line')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Gradient Descent Linear Fit')
    plt.legend()
    plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x, y)
```
